```python
def crear_columnas(file_path, id, column_names, df_final):
    # Cargar el archivo CSV en un DataFrame, especificando las columnas
    df = pd.read_csv(file_path, header=None, names=column_names)
    
    # Añadir la columna 'id' con el valor proporcionado
    df['id'] = id
    
    # Reorganizar las columnas para que 'id' esté al final
    cols = list(df.columns)  # Obtener la lista de todas las columnas
    cols.append(cols.pop(cols.index('id')))  # Mover la columna 'id' al final
    df = df[cols]  # Reorganizar el DataFrame con las columnas en el nuevo orden

    # Combinar el nuevo DataFrame con el DataFrame final
    df_combined = pd.concat([df_final, df], ignore_index=True)
    
    return df_combined  # Retornar el DataFrame combinado

# Definir los nombres de las columnas que se espera en el CSV
column_names = ['instruction', 'option_a', 'option_b', 'option_c', 'option_d', 'answer']  # Cambia esto según tus necesidades

# Crear un DataFrame final vacío con las columnas especificadas
df_final = pd.DataFrame(columns=column_names + ['id'])  # Asegúrate de que tenga las columnas correctas

# Rutas a los archivos CSV
file_path1 = './data/test/high_school_world_history_test.csv'
file_path2 = './data/test/high_school_us_history_test.csv'
file_path3 = './data/test/high_school_european_history_test.csv'
# file_path4 = './data/test/high_school_government_and_politics_test.csv' 

# Llamar a la función para cada archivo y combinar los resultados
df_final = crear_columnas(file_path1, 'high_school_world_history', column_names, df_final)
df_final = crear_columnas(file_path2, 'high_school_us_history', column_names, df_final)
df_final = crear_columnas(file_path3, 'high_school_european_history', column_names, df_final)
# df_final = crear_columnas(file_path4, 'high_school_government_and_politics', column_names, df_final)

# Mostrar el DataFrame combinado
print(df_final.head())
```

```python
# PARA OBTENER TODAS LAS PREGUNTAS DEL MMLU EN UN SOLO ARCHIVO Y CON LA CATEGORÍA
import pandas as pd
import os

# Ruta de la carpeta que contiene los archivos CSV
folder_path = './data/test/'

# Nombres de las columnas
column_names = ['instruction', 'option_a', 'option_b', 'option_c', 'option_d', 'answer']

# Lista para almacenar los DataFrames
dataframes = []

# Iterar sobre todos los archivos CSV en la carpeta
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        # Leer el archivo CSV
        df = pd.read_csv(file_path, header=None, names=column_names)
        # Añadir una columna con el nombre del archivo (sin la extensión .csv)
        df['id'] = os.path.splitext(filename)[0]
        # Añadir el DataFrame a la lista
        dataframes.append(df)

# Concatenar todos los DataFrames en uno solo
combined_df = pd.concat(dataframes, ignore_index=True)

# Guardar el DataFrame combinado en un archivo Excel
output_file = "MMLU_completo.xlsx"
combined_df.to_excel(output_file, sheet_name='en',index=False)

print(f"Dataset combinado guardado en '{output_file}'.")
```

```python
import pandas as pd
import ast
import re

# Cargar el archivo Excel original
file_path = "MMLU_pro.xlsx"  # Reemplázalo con la ruta correcta
sheet_name = "Sheet1"  # Ajusta si el nombre de la hoja es diferente

# Leer el archivo Excel
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Función para extraer opciones correctamente
MAX_OPTIONS = 10  # Hasta la opción "j"
def extract_options(options):
    try:
        options_list = re.findall(r"'([^']+)'", options)  # Extrae el texto dentro de las comillas simples
        if not options_list:
            return None  # Si no hay opciones extraídas, descartar
        return options_list[:MAX_OPTIONS]  # Limitar a 10 opciones (a-j)
    except Exception:
        return None  # Si hay error, descartar

# Aplicar extracción de opciones
df["options"] = df["options"].apply(lambda x: extract_options(str(x)))

# Eliminar filas con errores
df = df.dropna(subset=["options"])

# Crear nuevas columnas para cada opción (máximo 10: a-j) y rellenar con "-" si hay menos
df_options = df["options"].apply(pd.Series)
df_options.columns = [f"option_{chr(97 + i)}" for i in range(MAX_OPTIONS)]
df_options = df_options.fillna("-")  # Rellenar opciones faltantes con "-"

# Unir con el dataframe original sin la columna "options"
df_final = df.drop(columns=["options"]).join(df_options)

# Guardar en un nuevo archivo Excel
output_path = "MMLU_pro_separado.xlsx"
df_final.to_excel(output_path, index=False)

print(f"Archivo guardado en: {output_path}")
```

```python
import pandas as pd
import ast
import re

# Cargar el archivo Excel original
file_path = "MMLU_pro.xlsx"  # Reemplázalo con la ruta correcta
sheet_name = "Sheet1"  # Ajusta si el nombre de la hoja es diferente

# Leer el archivo Excel
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Función para extraer opciones correctamente
MAX_OPTIONS = 10  # Hasta la opción "j"
def extract_options(options):
    try:
        options_list = re.findall(r"'([^']+)'", options)  # Extrae el texto dentro de las comillas simples
        if not options_list:
            return None  # Si no hay opciones extraídas, descartar
        return options_list[:MAX_OPTIONS]  # Limitar a 10 opciones (a-j)
    except Exception:
        return None  # Si hay error, descartar

# Aplicar extracción de opciones
df["options"] = df["options"].apply(lambda x: extract_options(str(x)))

# Eliminar filas con errores
df = df.dropna(subset=["options"])

# Crear nuevas columnas para cada opción (máximo 10: a-j) y rellenar con "-" si hay menos
df_options = df["options"].apply(pd.Series)
df_options.columns = [f"option_{chr(97 + i)}" for i in range(MAX_OPTIONS)]
df_options = df_options.fillna("-")  # Rellenar opciones faltantes con "-"

# Crear nuevo DataFrame con la estructura deseada
df_final = pd.DataFrame()
df_final["instruction"] = df["question"]  # Pregunta como instrucción
df_final[[f"option_{chr(97 + i)}" for i in range(MAX_OPTIONS)]] = df_options  # Opciones
df_final["answer"] = df["answer"]  # Respuesta correcta
df_final["id"] = df["category"]  # Categoría

# Guardar en un nuevo archivo Excel
output_path = "MMLU_pro_completo.xlsx"
df_final.to_excel(output_path, index=False, sheet_name='en')

print(f"Archivo guardado en: {output_path}")
```
